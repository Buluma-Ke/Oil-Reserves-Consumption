## *Reshape DataFrame for analysis*

```python
pd.options.display.float_format = "{:,.4f}".format  # 4 dp
rsrv_df['World Crude Oil Reserves (billion)'] = rsrv_df['World Crude Oil Reserves (billion)'].str.replace(' ', '_')
rsrv_df.iloc[:, 1:] = rsrv_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
rsrv_df.replace(0, np.nan, inplace=True) 
rsrv_df.dropna(how='any', inplace=True) #drop null values

rsrv_df.set_index('World Crude Oil Reserves (billion)', inplace=True)
rsrv_df = rsrv_df.T     # Transpose the df
rsrv_df.reset_index(inplace=True)

rsrv_df.rename(columns={'index': 'Year'}, inplace=True)
rsrv_df['Year'] = pd.to_datetime(rsrv_df['Year'], format='%Y')
rsrv_df['Year'] = rsrv_df['Year'].dt.year
rsrv_df.set_index('Year', inplace=True)

print(rsrv_df.shape)
rsrv_df.sample(10)
```

# **DATA VISUALISATION**
### Global oil reserves 1995- 2021

```python
plt.figure(figsize=(14, 7))
sns.set_style("darkgrid")
sns.lineplot(data=rsrv_df['World'])

def hide_ticks():
    for i, l in enumerate(plt.gca().xaxis.get_ticklabels()):
        val = int(l.get_text())
        if val % 2 != 0:
            l.set_visible(False)

hide_ticks() 

plt.title('Global oil reserves from 1995 - 2021')
plt.ylabel('Year')
plt.ylabel('Crude Oil Reserves (billion)')
plt.show()
```

## World largest oil reserves by country

```python
# Top 10 countries with the largest oil reserves
def plot_top_reserves(year):
    i = rsrv_df.index.get_loc(year)
    Top_10 = rsrv_df.iloc[i, 1:78].sort_values(ascending=False)
    top_10 = Top_10.head(10) # Top 10
        
    sns.set_style("darkgrid")
    plt.figure(figsize=(15, 7))
    columns = top_10.index
    values = top_10.values

    sns.barplot(x=values, y=columns, hue=columns, palette= "dark:red", edgecolor='black')
    plt.title(f'Top 10 countries with the largest oil reserves({year})')
    plt.ylabel('Country')
    plt.xlabel('Crude Oil Reserves (billion)')

    #add values beside bar plot
    for i, value in enumerate(values):
        plt.text(value + 0.1, i, str(value), va='center')
        
    plt.tight_layout()
    plt.show()
    
# 1995
plot_top_reserves(1995)
```

```python
# Grow rate of the top 10 oil reserves
def plot_growth_rate(year):
    i = rsrv_df.index.get_loc(year)
    Top_10 = rsrv_df.iloc[i, 1:78].sort_values(ascending=False)
    top_10 = Top_10.head(10) # Top 10
            
    sns.set_style("darkgrid")
    plt.figure(figsize=(15, 7))
    columns = top_10.index
    values = top_10.values

    sns.lineplot(data=rsrv_df[columns])
    plt.xlabel('Year')
    plt.ylabel('Crude oil reserves (billions)')
    plt.title(f'GROWTH RATE OF {year} TOP 10 OIL RESERVES')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.show()
    
#1995
plot_growth_rate(1995)
```
